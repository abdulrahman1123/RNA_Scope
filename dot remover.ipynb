```python
import os
from pathlib import Path
import numpy as np

# The base directory. Change this line to your parent directory (the one that includes tif files directly, e.g. NF200 or P2RX ...)
base_dir = Path("/Volumes/Maria_3/MASTER THESIS_MARIA/Predictions_DS8/P2RX3")
tif_files = [item for item in base_dir.iterdir() if str(item).endswith('.tif') or str(item).endswith('.tiff')]
for f_dir in tif_files:
    fname = str(f_dir.name)
    dot_inds = np.where(np.array([chr=='.' for chr in fname]))[0]
    if len(dot_inds) >1:
        for d_ind in dot_inds[0:-1]:
            fname = fname[:d_ind] + '_' + fname[d_ind + 1:]
    new_f_dir = f_dir.parent/fname
    os.rename(str(f_dir),str(new_f_dir))
    print(f'The file \033[34m{str(f_dir.name)}\033[0m was renamed to \033[34m{str(fname)}\033[0m')


```
